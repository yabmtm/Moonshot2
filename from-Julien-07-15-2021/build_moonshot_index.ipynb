```python
import pandas as pd
df = pd.read_csv('JUL-TUD-06b2044f.csv')
df
```

```python
sdf_files = ['Mpro-x11294_2.sdf', 'Mpro-x12692_DMSO_removed.sdf', 'Mpro-P0009_2.sdf']
smiles = {} # smiles[sdf_file] = []  <-- ordered list of SMILES string for RUN0, RUN1, etc...

for sdf_file in sdf_files:

    smiles[sdf_file] = []
    
    # read in all the SMILES strings
    fin = open(sdf_file, 'r')
    lines = fin.readlines()
    fin.close()
    
    for i in range(len(lines)):
        if lines[i].count('> <Smiles>') > 0:
            smiles[sdf_file].append(lines[i+1].strip())
    n = len(smiles[sdf_file])
    print(f'Found {n} SMILES for {sdf_file}')
# print('smiles', smiles)


# For each SMILES string found, look up the Postera Link from the csv table above
for sdf_file in sdf_files:

    for run in range(len(smiles[sdf_file])):
        smile_string = smiles[sdf_file][run]
    
        entry = df[df['SMILES'] == smile_string ]
        try:
            CID_string = df.loc[df['SMILES'] == smile_string, 'CID (canonical)'].iloc[0]
            Postera_Link = df.loc[df['SMILES'] == smile_string, 'Postera Link'].iloc[0]
            print(f'| {sdf_file} | RUN{run} | <a href="{Postera_Link}">{CID_string}</a> |' )
        except:
            print(f'| {sdf_file} | RUN{run} | Postera Entry not found: {smile_string} |')
            # print(entry)
        
```
